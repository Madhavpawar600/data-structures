# Linkedlist class and constructor

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class Linkedlist:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
        
    def getnodeat(self,idx):
        temp=self.head
        for i in range(0,idx):
            temp=temp.next
        return temp
        
    def addfirst(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            new_node.next=self.head
            self.head=new_node
        self.size+=1
            
    def addlast(self,data):
        new_node=Node(data)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            self.tail=new_node
        self.size+=1
            
            
    def removefirst(self):
        if self.size==0:
            print('linkedlist is empty')
            return
        elif self.size==1:
            self.head=self.tail=None
            self.size-=1
        else:
            self.head=self.head.next
            self.size-=1
                
            
    def removelast(self):
        if self.size==0:
            print('linkedlist is empty')
            return
        elif self.size==1:
            self.head=self.tail=None
            self.size-=1
            return
        else:
            temp=self.head
            for i in range(0,self.size-2):
                temp=temp.next
            self.tail=temp
            self.tail.next=None
            self.size-=1
                
    def display(self):
        temp=self.head
        while temp!=None:
            print(temp.data,end=' ')
            temp=temp.next
    
        
                
l1=Linkedlist()
l1.addfirst(2)
l1.addfirst(1)
l1.addlast(3)
l1.addlast(4)
print(l1.size)
print(l1.head.data)
print(l1.tail.data)
l1.display()
l1.removelast()
l1.addfirst(4)
print('\nupdated list')
l1.display()
l1.size
```

# Even ODD

```python
def even_odd(llist):
    even=Linkedlist()
    odd=Linkedlist()
    
    while llist.size>0:
        temp=llist.head.data
        llist.removefirst()
        if temp%2==0:
            even.addlast(temp)
        else:
            odd.addlast(temp)
    
    odd.tail.next=even.head
    llist.head=odd.head
    llist.tail=even.tail
    llist.size=even.size+odd.size
    del even,odd
even_odd(newlist)
```

# Middle of linkedlist 

```python
def middle(llist):
    slow=llist.head
    fast=llist.head.next
    
    while fast.next and fast.next.next:
        slow=slow.next
        fast=fast.next.next
    middle=slow
    return middle
middle(newlist)
```

# Reverse linkedlist data iterative 

```python
def reverse_iter(llist):
    l=0
    r=llist.size-1
    while l < r:
        node1=llist.getnodeat(l)
        node2=llist.getnodeat(r)
        temp=node1.data
        node1.data=node2.data
        node2.data=temp
        l+=1
        r-=1
reverse_iter(newlist)
```

# Reverse linkedlist pointer iterative 

```python
#basically we change the next pointer to last node
def reverse_list(llist):
    temp=llist.head
    prev=None
    next=None
    
    while temp!=None:
        next=temp.next
        temp.next=prev
        prev=temp
        temp=next
    llist.head=prev
reverse_list(newlist)
```

# Reverse a linkedlist pointer recursive

```python
def rec_reverse_helper(head,tail):
    if head==None:
        return
    rec_reverse_helper(head.next,tail)
    if head==tail:
        pass
    else:
        head.next.next=head
        
def rec_reverse(llist):
    rec_reverse_helper(llist.head,llist.tail)
    llist.head.next=None
    temp=llist.head
    llist.head=llist.tail
    llist.tail=temp
rec_reverse(newlist)
```

# Reverse k groups list 

```python
def reversekgrp(head,k):
    temp=head
    prev=None
    next=None
    count=0
    while temp is not None and count<k:
        next=temp.next
        temp.next=prev
        prev=temp
        temp=next
        count+=1
    if next!=None:
        head.next=reversekgrp(next,k)
    return prev
newlist.head=reversekgrp(newlist.head,3)
```

# Reverse k group list another approach and if group!=k not reverse

```python
def reverse_kgrp(llist,k):
    prev=Linkedlist()
    
    while llist.size!=0:
        curr=Linkedlist()
        if llist.size>=k:
            for i in range(0,k):
                temp=llist.head.data
                llist.removefirst()
                curr.addfirst(temp)
        else:
            s=llist.size
            temp=llist.head.data
            for i in range(0,s):
                llist.removefirst()
                curr.addlast(temp)
                
        if prev.head==None:
            prev=curr
        else:
            prev.tail.next=curr.head
            prev.tail=curr.tail
            prev.size+=curr.size
            
    llist.head=prev.head
    llist.tail=prev.tail
    llist.size=prev.size
    del prev
reverse_kgrp(newlist,2)
```

# Kth node from end of linked list

```python
def kthnode(linkedlist,k):
    count=0
    temp=linkedlist.head
    kthnode=linkedlist.head
    
    while count<k and temp!=None:
        temp=temp.next
        count+=1
    while temp!=None:
        temp=temp.next
        kthnode=kthnode.next
    return kthnode.data

kthnode(newlist,3)
```

# Detect loop in linked list

```python
#create list with loop
t1=Node(1)
t1.next=Node(2)
t1.next.next=Node(3)
t1.next.next.next=Node(4)
t1.next.next.next.next=Node(5)
t1.next.next.next.next=t1.next.next.next
```

```python
def detectloop(linkedlist):
    slow=linkedlist
    fast=linkedlist.next
    
    while fast.next and fast.next.next:
        if slow==fast:
            print('loop detected')
            return
        slow=slow.next
        fast=fast.next.next
        
    print('no loop detected')
detectloop(t1)
```

# merge two sortedlist 

```python
first_sortedlist=create_list([1,3,5,7,9])
second_sortedlist=create_list([2,4,6,8,10])
```

```python
def mergetwolist(first,second):
    if first.head==None:
        return second
    if second.head==None:
        return first
    temp=Linkedlist()
    while first.size!=0 and second.size!=0:
        if first.head.data<second.head.data:
            val=first.head.data
            first.removefirst()
            temp.addlast(val)
        else:
            val=second.head.data
            second.removefirst()
            temp.addlast(val)
    
    while first.size!=0:
        val=first.head.data
        first.removefirst()
        temp.addlast(val)
    while second.size!=0:
        val=second.head.data
        second.removefirst()
        temp.addlast(val)
    return temp
sortedlist=mergetwolist(first_sortedlist,second_sortedlist)
```

# Merge Sort

```python
def middleformerge(head,tail):
    slow=head
    fast=head
    while fast.next!=tail and fast.next.next!=tail:
        slow=slow.next
        fast=fast.next.next
    return slow
```

```python
def mergesort(head,tail):
    
    if head==tail:
        bs=Linkedlist()
        bs.addfirst(head.data)
        return bs
    
    mid=middleformerge(head,tail)
    first=mergesort(head,mid)
    second=mergesort(mid.next,tail)
    return mergetwolist(first,second)
```
